## Info globale

```python
# Pour chaque name on a un dictionnaire de performances

def table(frame):
    
    df1 = pd.DataFrame([])
    
    for name in frame['name']:
        try : 
            dico = {}
            # configure url 
            time.sleep(1)
            url = frame.loc[frame.name == name,'link'].values[0]
            response = requests.get(url)
            season_page = BeautifulSoup(response.content)

            # configure generale info 
            globale = season_page.find('div',attrs={'id' :'left-panel'})
            gl_list_info = globale.findAll('div',attrs={'class' :'s-list-info'})

            # scrapping 
            for h in range(len(gl_list_info)):
                for i in range(len(gl_list_info[0].findAll('div',attrs={'class':'s-list-info__value'}))):
                    label = gl_list_info[h].findAll('div',attrs={'class':'s-list-info__label'})[i].text.strip().strip(':')
                    value = gl_list_info[h].findAll('div',attrs={'class':'s-list-info__value'})[i].text.strip().strip(':')

                    # add to dico
                    dico[str(label)] = value
                    print(label +' : ' + value)
            
            # price of the suscribe 
            btn_price = season_page.find('div',attrs={'class' :'s-btns-area'})
            price = btn_price.text.replace('\xa0',' ').replace('\n',' ').replace('Copy -','').strip()
            
            
            df2 = pd.DataFrame([dico])
            df2.loc[0,'name'] = name
            df2.loc[0,'last_refresh'] = pd.to_datetime('now')
            df2.loc[0,'price'] = price 
            
            
            df1 = pd.concat([df1,df2]).reset_index(drop=True)
            
            # conclusion
            print(f'Dataframe Successfully updated :)! for {name}')
            print('\n')
        except :
            print(f'ERROR :( for {name}')
            
        
    return df1
```

```python
glb = table(frame)
glb.columns = [i.replace(' ','') for i in glb.columns]
```

## Drawdown

> The Balance reflects your profit/loss from closed positions.
The Equity reflects the real-time calculation of your profit/loss. The Equity takes into account both open AND closed positions.


```python
drawdown = pd.DataFrame([])

for name in frame['name']:
    time.sleep(1)
    
    url = frame.loc[frame.name == name,'link'].values[0]
    try : 
        print(url)
        response = requests.get(url)
        print(response)
        season_page = BeautifulSoup(response.content)

        globale = season_page.find('div',attrs={'class' :"s-data-columns hoverable"})
        f1 = globale.findAll('div',attrs={'class':'s-data-columns__column'})[-1]
        f2 = f1.findAll('div',attrs={'class':'s-data-columns__value'})[-2].text
        f3 = f1.findAll('div',attrs={'class':'s-data-columns__value'})[-1].text
        
        df = pd.DataFrame([{'name':name ,'max_drawdown_balance':float(f2[:4]),'max_drawdown_equity':float(f3[:4]), 'url':url}])

        drawdown = pd.concat([drawdown,df]).reset_index(drop=True)
    except : 
        print(response)


drawdown.name = drawdown.name.astype(str)
```

## Growth

```python
def fmt(perf):
    res = pd.DataFrame.from_dict(perf, orient='index').reset_index()
    res['index'] = res['index'].apply(lambda x: x.replace('Growth for','').replace('.','-').replace(':','').strip())
    res[0] = res[0].apply(lambda x: x.replace('%','').replace(':','').strip()).astype(float)
    # Filter unpertinent values
    res['verif'] = np.where(res['index'].isin(["2021","2022","2019"]),1,0)
    res = res[res.verif ==0]
    
    res.columns = ['month-year','rent','verif']
    res['month-year'] = pd.to_datetime(res['month-year']).dt.to_period('M')
    res = res.drop(['verif'],axis=1)
    
    return res


def grow(res):
    all_time = {}
    all_time['all_rent_sup10'] = (res.rent > 10).sum()
    all_time['all_rent_prob']  = all_time['all_rent_sup10']/len(res.rent)
    all_time['all_rent_mean'] = res.rent.mean()
    all_time['all_rent_q1'] = res.rent.quantile(0.25)
    all_time['all_rent_median'] = res.rent.median()
    all_time['all_rent_q3'] = res.rent.quantile(0.75)
    all_time['all_rent_std'] = res.rent.std()
    all_time['all_rent_min'] = res.rent.min()
    all_time['all_rent_max'] = res.rent.max()

    last_6months = {}
    last_6months['l6_rent_sup10'] = (res[-6:].rent > 10).sum()
    last_6months['l6_rent_prob']= last_6months['l6_rent_sup10']/6
    last_6months['l6_rent_mean'] = res[-6:].rent.mean()
    last_6months['l6_rent_q1'] = res[-6:].rent.quantile(0.25)
    last_6months['l6_rent_median'] = res[-6:].rent.median()
    last_6months['l6_rent_q3'] = res[-6:].rent.quantile(0.75)
    last_6months['l6_rent_std'] = res[-6:].rent.std()
    last_6months['l6_rent_min'] = res[-6:].rent.min()
    last_6months['l6_rent_max'] = res[-6:].rent.max()
    
    dataframe = pd.DataFrame([last_6months]).join(pd.DataFrame([all_time]))
    
    return dataframe
```

```python
growth = pd.DataFrame([])

for name in frame['name']:
    # 1 second of rest
    time.sleep(1)

    url = frame.loc[frame.name == name,'link'].values[0]
    response = requests.get(url)
    season_page = BeautifulSoup(response.content)

    print(response)

    globale = season_page.find('div',attrs={'id' :"returnChart"})

    perf = {}

    # fill the dictionary with growth data 
    try : 
        for i in range(1,len(globale.findAll('div',attrs={'class' :"row"}))-1):
            print('-'*50)           
            perf_page = globale.findAll('div',attrs={'class' :"row"})[i]
            year = perf_page.find('div',attrs={'class',"cell year"}).text

            print(year)
            for i in perf_page.findAll('div',attrs={'class',"cell"})[1:]:
                try : 
                    perf[str(i['title'][:-7])]= str(i['title'][-7:])
                except : 
                    print('no data yet :(')
                    print('-'*50) 

        res = fmt(perf)
        df1 = grow(res)

        df1['name'] = name
        df1['link'] = url 

        growth = pd.concat([growth,df1]).reset_index(drop=True)
    
    except:
        pass
             
```
