```python
def table(frame):
    
    df1 = pd.DataFrame([])
    
    i = 1
    for name in keep_scrap :
        if i%50 == 0:
            time.sleep(random.randrange(1,100))
            
        dico = {}
        # configure url 
        time.sleep(random.randrange(1,5))
        url = frame.loc[frame.name == name,'link'].values[0]
        print(url)
        response = requests.get(url)
        season_page = BeautifulSoup(response.content)

        # configure generale info 
        globale = season_page.find('div',attrs={'id' :'left-panel'})
        gl_list_info = globale.findAll('div',attrs={'class' :'s-list-info'})

        # scrapping 
        for h in range(len(gl_list_info)):
            for i in range(len(gl_list_info[0].findAll('div',attrs={'class':'s-list-info__value'}))):
                label = gl_list_info[h].findAll('div',attrs={'class':'s-list-info__label'})[i].text.strip().strip(':')
                value = gl_list_info[h].findAll('div',attrs={'class':'s-list-info__value'})[i].text.strip().strip(':')

                # add to dico
                dico[str(label)] = value
                print(label +' : ' + value)
            
         # price of the suscribe 
        btn_price = season_page.find('div',attrs={'class' :'s-btns-area'})
        price = btn_price.text.replace('\xa0',' ').replace('\n',' ').replace('Copy -','').strip()
            
            
        df2 = pd.DataFrame([dico])
        df2.loc[0,'name'] = name
        df2.loc[0,'last_refresh'] = pd.to_datetime('now')
        df2.loc[0,'price'] = price 
            
            
        df1 = pd.concat([df1,df2]).reset_index(drop=True)
            
        # conclusion
        print(f'Dataframe Successfully updated :)! for {name}')
        print(i)
        print('\n')
            
        i+=1
        
            #except :
            #  print(f'ERROR :( for {name}')
    return df1
```

```python
df = table(frame)
```
